```python
import matplotlib.pyplot as plt
import yfinance as yf
from matplotlib import style
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
from mpl_finance import candlestick_ohlc
style.use('fivethirtyeight')
MA1 = 10
MA2 = 30
def moving_average(values,window):
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values, weights, 'valid')
    return smas
def high_minus_low(highs,lows):
    return highs-lows
def graph_data(stock):
    fig = plt.figure()
    stock_data = yf.download(stock, period = '1y', interval = '1d')
    print(stock_data)
    stock_data = stock_data.reset_index()
    print(stock_data)
    stock_data['Date'] = stock_data['Date'].apply(mdates.date2num)
    start = len(stock_data.Date[MA2-1:])
    ax1 = plt.subplot2grid((6,1),(0,0), rowspan = 1, colspan = 1)
    plt.title(stock)
    plt.ylabel('High-Low')
    ax2 = plt.subplot2grid((6,1),(1,0), rowspan = 4, colspan = 1, sharex = ax1)
    plt.ylabel('Price')
    ax2v = ax2.twinx()
    ax3 = plt.subplot2grid((6,1),(5,0), rowspan = 1, colspan = 1, sharex = ax1)
    plt.ylabel('Mvng. avg.')
    ax1.xaxis_date()
    ax2.xaxis_date()
    ax3.xaxis_date()
    candlestick_ohlc(ax2, stock_data.values[-start:], width = 0.1)
    h_l = list(map(high_minus_low, stock_data['High'], stock_data['Low']))
    ma1 = moving_average(stock_data['Close'],MA1)
    ma2 = moving_average(stock_data['Close'],MA2)
    start = len(stock_data.Date[MA2-1:])
    bbox_props = dict(boxstyle = 'round4', fc = 'w', ec = 'k', lw = 1)
    ax2.annotate(str(stock_data.iloc[-1]['Close']), (stock_data.Date.tail(1),stock_data.Close.tail(1)), xytext = (stock_data.Date.tail(1)+20, stock_data.Close.tail(1)), bbox = bbox_props, arrowprops = dict(color = 'grey'))
    ax3.plot(stock_data.Date[-start:],ma1[-start:], color = 'k', lw = 1.5, label = str(MA1)+'MA')
    ax3.plot(stock_data.Date[-start:],ma2[-start:], color = 'k', lw = 1.5, label = str(MA2)+'MA')
    ax3.fill_between(stock_data.Date[-start:], ma1[-start:], ma2[-start:], where = ma1[-start:]<=ma2[-start:], color = 'r', alpha = 0.25)
    ax3.fill_between(stock_data.Date[-start:], ma1[-start:], ma2[-start:], where = ma1[-start:]>ma2[-start:], color = 'g', alpha = 0.25)
    ax1.plot(stock_data.Date[-start:], h_l[-start:], label = 'H-L', color = 'k', lw = 1)
    ax2v.fill_between(stock_data.Date[-start:], stock_data.Volume[-start:],0,color = 'r', alpha = 0.25)
    ax2v.plot([],[], color = 'r', alpha = 0.25, label = 'Volume')
    ax2v.axes.yaxis.set_ticklabels([])
    ax2v.grid(False)
    ax2v.set_ylim(0,3*stock_data.Volume.max())
    plt.setp(ax1.get_xticklabels(), visible = False)
    plt.setp(ax2.get_xticklabels(), visible = False)
    plt.subplots_adjust(left = 2,right = 5,top = 5,bottom =2)
    ax1.legend()
    ax2v.legend()
    ax3.legend()
    plt.show()
    fig.savefig('GOOGL.png')
graph_data('GOOGL')
```
