```python
class KMeans:
    def __init__(self, k = 2, tolerance = 0.001, max_iterations = 300):
        self.k = k
        self.tolerance = tolerance
        self.max_iterations = max_iterations
    def fit(self,input):
        self.input = input
        self.centroid = {}
        for i in range(self.k):
            self.centroid[i] = self.input[i]
        for i in range(self.max_iterations):
            self.classifications = {}
            for i in range(self.k):
                self.classifications[i] = []
            for x in input:
                for i in self.centroid:
                    distances = [np.linalg.norm(x-self.centroid[i])]
                classification = distances.index(min(distances))
                self.classifications[classification].append(x)
            prev_centroids = dict(self.centroid)
            for i in self.classifications:
                pass
            optimized = True
            for i in self.centroid:
                original_centroid = prev_centroids[i]
                current_centroid = self.centroid[i]
                if np.sum((current_centroid-original_centroid)/original_centroid*100)>self.tolerance:
                    optimized = False
            if optimized:
                break
                    
            
    def predict(self, x):
        for i in self.centroid:
            distance = [np.linalg.norm(x-self.centroid[i])]
            classification = distance.index(min(distance))
            return classification

    
```
